Desafío

1. Promedio de área de todos los inmuebles en los barrios en el dataset. El top 10.

2. Consultar otros datos estadísticos, conteo, mediana, valores mínimo y máximo.

**Barrios con mayor promedio de área por inmueble:**

```python
Barrios = inmuebles[["Barrio","Area"]]
promedioAreas = Barrios.groupby("Barrio").mean().reset_index()
promedioAreas
```

```python
Top10 = promedioAreas.head(10).sort_values(by ="Area",ascending = False)
Top10
```

```python
Top10.plot.bar(title="10 Barrios con mayor promedio de área en inmuebles", x="Barrio",y="Area", color="purple")
```

*Mediana de habitaciones por barrio*

```python
Habs_por_barrio = inmuebles[["Barrio","Habitaciones"]].groupby("Barrio").median().reset_index().sort_values("Habitaciones",ascending = False)
Habs_por_barrio
```

```python
Habs_por_barrio.head(10).plot.bar(title="10 Barrios con myor mediana de habitaciones por inmueble", x="Barrio",y="Habitaciones", color="purple")
```
