**Barrios con mayor promedio de área por inmueble:**

```python
Barrios = inmuebles[["Barrio","Area"]]
promedioAreas = Barrios.groupby("Barrio").mean().reset_index()
promedioAreas
```

```python
Top10 = promedioAreas.head(10).sort_values(by ="Area",ascending = False)
Top10
```

```python
Top10.plot.bar(title="10 Barrios con mayor promedio de área en inmuebles", x="Barrio",y="Area", color="purple")
```

```python
valor = inmuebles.Valor.str.split(expand=True)
inmuebles['Moneda'] = valor[0]
inmuebles['Precio'] = valor[1]
inmuebles.sample
```

```python
inmuebles.info()
```

```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','',regex=True)
```

```python
inmuebles[['Precio','Barrio']]
```

```python
inmuebles.info()
```

```python
inmuebles['Precio_Millon'] = inmuebles.Precio.astype('float')/1000000
inmuebles.info()
```

```python
inmuebles.describe()
```

```python
pd.set_option('display.precision',2)
pd.set_option('display.float_format',lambda x:'%.2f' % x)
inmuebles.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))

grafica = sns.histplot(data=inmuebles, x='Precio_Millon', kde=True, hue='Tipo')
grafica.set_title('Distribución de valor de los inmuebles en Bógota')
plt.xlim(50,1000)
# plt.ylim(0,8)
plt.savefig('/content/drive/MyDrive/Inmersion Ciencia de Datos de Alura/exports/valor_inmuebles.png',format='png')
plt.show()
```

*Precio del m2 por barrio*

```python
inmuebles['Precio_m2'] =inmuebles.Precio_Millon/inmuebles.Area
inmuebles.Precio_m2
inmuebles.info()
```

```python
Precio_m2_barrio = inmuebles[['Barrio','Precio_m2']].groupby("Barrio").median().head(10).reset_index().sort_values("Precio_m2",ascending = False)
Precio_m2_barrio
```

```python
plt.figure(figsize=(8,6))

grafica = sns.barplot(data=Precio_m2_barrio, x='Barrio', y='Precio_m2')
grafica.set_title('Precio del m2 por barrio')

plt.show()
```
